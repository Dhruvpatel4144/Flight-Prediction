# Feature Engineering Process

```python
# here my date journey is in object that is string format, i have to convert this into date-time format
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
# or you can use lambda function

# final_df['Date'] = final_df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
# final_df['Month'] = final_df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
# final_df['Year'] = final_df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
final_df.head()
```

```python
# converting object datatype to integer

final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
# now i will drop the features 'Date_of_Journey'

final_df.drop('Date_of_Journey', axis=1, inplace=True) 
```

```python
final_df.head()
```

```python
# Another feature 'Arrival_Time'

final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
# now for 'Duration'

final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
# convert object datatype into integer

final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
# for 'Dep_Time' split, change datatype and drop

final_df['Dept_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dept_min'] = final_df['Dep_Time'].str.split(':').str[1]

final_df['Dept_hour'] = final_df['Dept_hour'].astype(int)
final_df['Dept_min'] = final_df['Dept_min'].astype(int)

final_df.drop('Dep_Time', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
# another feature 'Total_Stops'
final_df['Total_Stops'].unique()
```

```python
# by using map function                                                                                                                                         

final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'NaN':1 })
```

```python
# final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df.head()
```

```python
# drop 'Route'

final_df.drop('Route', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
# another feature 'Duration' 

# this is series so we need to split accordingly

final_df['duration_hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
final_df['duration_hour'] = final_df['duration_hour'].astype(int)
```

```python
# from above error ,there is '5m' value in duration_hour

final_df[final_df['duration_hour'] == '5m']

# how this value is '5m', because we can't reach mumbai to hydrabad in just '5m'
# now we have drop this record beacuse it doesn't make any sense
```

```python
# droping above values

final_df.drop(6474, axis=0, inplace = True)
final_df.drop(2660, axis=0, inplace = True)
```

```python
# now its deleted so we are getting empty values

final_df[final_df['duration_hour'] == '5m']     
```

```python
final_df['duration_hour'] = final_df['duration_hour'].astype(int)
```

```python
# drop 'Duration'
final_df.drop('Duration', axis=1, inplace = True)
```

```python
final_df.head()
```

```python
final_df.info()
```

```python
# rest all are categorical feature
# anouther feature  'Airline'

final_df['Airline'].unique()
```

```python
# so there are so many airlines here, so we will try here label encoding here
# we can also do with get.dummies but since we try to work with train and test data so it is better to use the transform techniques         

from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
